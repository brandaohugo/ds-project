```python
import random
def random_hex():
        random_number = random.randint(0,16777215)
        hex_number = str(hex(random_number))
        hex_number ='#'+ hex_number[2:]
        return str(hex_number)
```

```python
# function to produce JSON data

lst = []
for name in df.name.unique():
    dictionary = dict(data=list(df[df['name']==name]['used_cores'].values),
             label=name,
             borderColor=random_hex(),
             fill='false'
            )
    lst.append(dictionary)

print(lst)
```

```python
sim_time = [i for i in range(len(lst[0]['data']))]
print(sim_time)
```
